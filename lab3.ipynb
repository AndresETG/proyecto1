1) ¿Cuántos Municipios tienen más de 100000 habitantes?

a) ¿Cuál es la segunda ciudad más poblada?

```python
poblacion = df[['Ciudad', 'Poblacion']].sort_values(by='Poblacion')
print(f"La segunda ciudad más poblada es {poblacion.iloc[-2]['Ciudad']}")
```

3) ¿Cuántos municipios de Extremadura tienen más de 5000 habitantes?

La región de Extremadura está compuesta por las provincias de Cáceres y Badajoz. Por lo que no encontraremos en el dataframe.

```python
extremadura = df[df['Poblacion'] > 5000]
```

```python
municipio = extremadura[(extremadura['Provincia'] == 'Cáceres') | (extremadura['Provincia'] == 'Badajoz')].value_counts().size
print(f'Hay {municipio} municipios en la región de Extremadura con más de 5,000 habitantes')
```

4) ¿Cuál es el municipio situado más al Norte? (Usar el valor de la coordenada "Y" que representa la latitud en grados). Proporcione también la provincia a la que pertenece y su población

```python
norte = df[['Municipio','Provincia', 'Poblacion', 'Y']].sort_values(by='Y', ascending=False)
norte = norte[['Municipio','Provincia', 'Poblacion']].iloc[0]
norte
print(f'El municipio {norte["Municipio"]} con población de {norte["Poblacion"]} dentro de la provincia de {norte["Provincia"]} se encuentra más al Norte')
```

5) Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
provincia = df[df['Provincia']== 'Granada']
provincia.groupby('Provincia').agg(
    media = ('Poblacion', np.mean),
    mediana = ('Poblacion', np.median),
    desviacion_est = ('Poblacion', np.std),
    maximo = ('Poblacion', np.max),
    minimo = ('Poblacion', np.min)
    )
```

6) Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
histograma = df.groupby('Provincia')

for provincia, filas in histograma:
    poblacion = filas['Poblacion']
    plt.hist(poblacion, color='green', edgecolor='black')
    plt.xlabel('Población')
    plt.ylabel('Número de municipios')
    plt.title(f'Histograma de población para {provincia}')
    plt.show()
```
